Customer Personality Analysis

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("imakash3011/customer-personality-analysis")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

file_path = os.path.join(path, "marketing_campaign.csv")
df = pd.read_csv(file_path, sep='\t')
df.to_csv('marketing_campaign.csv')

df.head(20)
```

```python
df.info()
print("Dataset shape:", df.shape)
```

```python
#Data Cleaning

#Handle missing values
missing_values= df.isnull().sum()
print(missing_values)
```

```python
#drop rows with any missing values
df.dropna(inplace=True)
```

```python
#remove duplicate values
df.drop_duplicates(inplace=True)
```

```python
#Standardize text values
df['Education']=df['Education'].str.strip().str.replace(" ","_").str.lower().str.title()
df['Marital_Status']=df['Marital_Status'].str.strip().str.replace(" ","_").str.lower().str.title()
```

```python
#Convert Date foemat
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y', errors='coerce')
df['Dt_Customer'] = df['Dt_Customer'].dt.strftime('%d-%m-%Y')
```

```python
#rename column headers
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]
```

```python
#check and fix data types
df['income']=pd.to_numeric(df['income'], errors='coerce').astype('float')
```

```python
#Final Clened dataset
print("cleaned dataset shape", df.shape)
df.head(20)
```

```python
#save cleaned dataset
df.to_csv('cleaned_marketing_campaign.csv', index= False)
print("Cleaned dataset saved as cleaned_marketing_campaign.csv")
```

```python
#summary of changes made
summary = {
    "missing_values_handled": missing_values.to_dict(),
    "duplicates_removed": True,
    "columns_renamed": df.columns.tolist(),
    "date_format_converted": 'Dt_Customer' in df.columns,
    "text_standardized_columns": ["education", "marital_status"],
    "data_types_fixed": True,
}

print("\nSummary of Cleaning:")
for key, value in summary.items():
    print(f"{key}: {value}")
```
